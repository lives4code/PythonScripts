```python
from bs4 import BeautifulSoup
import requests
import json
import csv
import os
homePath = "/home/jacob/Documents/College/seniorProject/spaces"
spaces = []
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry

os.chdir(homePath)
session = requests.Session()
retry = Retry(connect=3, backoff_factor=0.5)
adapter = HTTPAdapter(max_retries=retry)
session.mount('http://', adapter)
session.mount('https://', adapter)

def requestProposal(name, skip):
    url1 = "https://hub.snapshot.org/graphql?operationName=Proposals&query=query%20Proposals%20%7B%0A%20%20proposals(%0A%20%20%20%20first%3A%2050%2C%0A%20%20%20%20skip%3A%20"                                                                                                                      
    url2 = "%2C%0A%20%20%20%20where%3A%20%7B%0A%20%20%20%20%20%20space_in%3A%20%5B%22"
    url3 = "%22%5D%2C%0A%20%20%20%20%20%20state%3A%20%22closed%22%0A%20%20%20%20%7D%2C%0A%20%20%20%20orderBy%3A%20%22created%22%2C%0A%20%20%20%20orderDirection%3A%20desc%0A%20%20)%20%7B%0A%20%20%20%20id%0A%20%20%20%20title%0A%20%20%20%20body%0A%20%20%20%20choices%0A%20%20%20%20start%0A%20%20%20%20end%0A%20%20%20%20snapshot%0A%20%20%20%20state%0A%20%20%20%20author%0A%20%20%20%20space%20%7B%0A%20%20%20%20%20%20id%0A%20%20%20%20%20%20name%0A%20%20%20%20%7D%0A%20%20%7D%0A%7D%0A"
    return url1 + str(skip) + url2 + str(name) + url3

def requestVote(proposal, skip):
    url1 = "https://hub.snapshot.org/graphql?operationName=Votes&query=query%20Votes%20%7B%0A%20%20votes%20(%0A%20%20%20%20first%3A%2050%2C%20%0A%20%20%20%20skip%3A%20"
    url2 = "%2C%20%0A%20%20%20%20where%3A%20%7B%0A%20%20%20%20%20%20proposal%3A%20%22"
    url3 = "%22%0A%20%20%20%20%7D%0A%20%20)%20%7B%0A%20%20%20%20id%0A%20%20%20%20voter%0A%20%20%20%20created%0A%20%20%20%20choice%0A%20%20%20%20space%20%7B%0A%20%20%20%20%20%20id%0A%20%20%20%20%7D%0A%20%20%7D%0A%7D%0A"
    return url1 + str(skip) + url2 + str(proposal) + url3

def parseProposal(entry):
    row = []
    try:
        ide = entry.get('id')
        title = entry.get('title')
        body = entry.get('body')
        choices = entry.get('choices')
        start = entry.get('start')
        end = entry.get('end')
        snapshot = entry.get('snapshot')
        auth = entry.get('auth')
        space = entry.get('space')
        row = [ide, title, body, choices, start, end, snapshot, auth, space, row]
    except:
        row = ['Error:not found']
    return row
            
def parseVote(entry):
    row = []
    try:
        ide = entry.get('id')
        voter = entry.get('voter')
        created = entry.get('created')
        choice = entry.get('choice')
        space = entry.get('space')
        row = [ide, voter, created, choice, space]
    except Exception as e: 
        print(e)
        row = ["error"]
    return row

def getProposals(proposalsCSV):
    proposals = []
    with open(proposalsCSV) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter = ',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                line_count = line_count + 1
            else:
                ide = str(row[0])
                proposals.append(ide)
    return proposals

def getSpaces():
    spaces = []
    with open('snapshotSpaces.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                print("nothing")
                line_count += 1
            else:
                directory = str(row[1])
                sirectory = directory[:-4]
                spaces.append(directory)
                
    return spaces
spaces = []
spaces = getSpaces()
for count, space in enumerate(spaces):
    print(count)
    path = os.path.join(homePath, space)
    #for every space create a directory
    try:
        os.mkdir(path)
    except:
        print('path already made')
    ##create a csv file for the proposals for that space
    proposalsCSV = path + "/" + space + ".csv"
    with open(proposalsCSV, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        ## write header
        writer.writerow(['id', 'title', 'body', 'choices', 'start', 'end', 'snapshot', 'auth', 'space'])
        ## get proposals
        skip = 0
        page = session.get(requestProposal(space, skip))
        #print(requestProposal(space + ".eth", skip))
        soup = BeautifulSoup(page.content, 'html.parser')
        #print(soup)
        ## while there is still data to get
        while soup.prettify() != "{\"data\":{\"proposals\":[]}}\n":
            try:
                data = json.loads(soup.prettify(), strict=False)
                data = data['data']
                data = data['proposals']
                #print(data)
                row = []
                for proposal in data:
                    row = parseProposal(proposal)
                    #print(row)
                    writer.writerow(row)
            except:
                writer.writerow(["error"])
                print("error with " + space + " count:" + str(count))
            skip +=50
            page = session.get(requestProposal(space, skip))
            soup = BeautifulSoup(page.content, 'html.parser')
    ##open the csv file you just created and parse each Proposal
    proposalIds = getProposals(proposalsCSV)
    ## for each proposal make a voting csv
    for proposal in proposalIds:
        votesCSV = path + "/" + proposal + ".csv"
        with open(votesCSV, 'w', newline = '') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(['id', 'voter', 'created', 'choice', 'space'])
            skip = 0
            page = session.get(requestVote(proposal, skip))
            soup = BeautifulSoup(page.content, 'html.parser')
            while soup.prettify() != "{\"data\":{\"votes\":[]}}\n":
                try:
                    data = json.loads(soup.prettify(), strict = False)
                    data = data['data']
                    data = data['votes']
                    #print(data)
                    for entry in data:
                        row = parseVote(entry)
                        #print(row)
                        writer.writerow(row)    
                except:
                    writer.writerow(["error"])
                    print("error with proposal" + proposal)
                skip += 50
                page = session.get(requestVote(proposal, skip))
                soup = BeautifulSoup(page.content, 'html.parser')
            
        
```
