```python
df = pd.read_csv('projectFile.csv')
df.head()
```

```python
string = 'West Hollywood, CA (2021)37*'
def stripDollar(string):
    if(isinstance(string, str)):
        if(re.sub('\$', '', string)):
            string = re.sub('\$', '', string)
    return string
def stripString(string):
    if( isinstance(string, str)):
        if(re.sub('([0-9]{4})', '', string) != None):
            string = re.sub('([0-9]{4})', '', string).replace('()', '')
        if(re.sub('\*', '', string) != None):
            string = re.sub('\*', '', string)
        if(re.sub('(\d)', '', string) != None):
            string = re.sub('\d', '', string)
    return string
df['Locality'] = df['Locality'].apply(stripString)
df = df.applymap(stripDollar)

df.head()
```
